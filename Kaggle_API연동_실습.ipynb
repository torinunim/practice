```python
train.drop(train[(train['OverallQual']<4) & (train['SalePrice']>200000)].index, inplace=True)
train.drop(train[(train['OverallCond']<4) & (train['SalePrice']>200000)].index, inplace=True)
train.reset_index(drop=True, inplace=True)
print(train.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
(mu, sigma)=norm.fit(train['SalePrice'])
print("The value of mu before log transformation is:", mu) # 평균
print("The value of sigma before log transformation is:", sigma) # 표준편차

fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(train['SalePrice'], color='b', stat="probability")
ax.xaxis.grid(False) # x 축 grid 비활성화
ax.set(ylabel="Frequency")
ax.set(xlabel="SalePrice")
ax.set(title="SalePrice distribution")

plt.axvline(mu, color='r', linestyle='--') # 새로 점선 추가
plt.text(mu+1000, 0.11, 'Mean of SalePrice', rotation=0, color='r')
fig.show()
```

norm.fit() 함수는 데이터 정규분포를 적합(fit)하는데 사용
정규 분포의 평균(mu)과 표준 편차(sigma)를 반환

```python
import numpy as np
train["SalePrice"]=np.log1p(train["SalePrice"]) # 배열 또는 시리즈의 각 요소에 대해 자연 로그 게산한 후 1을 더함

(mu, sigma)=norm.fit(train['SalePrice'])
print("The value of mu before log transformation is:", mu) # 평균
print("The value of sigma before log transformation is:", sigma) # 표준편차

fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(train['SalePrice'], color='b', stat="probability")
ax.xaxis.grid(False) # x 축 grid 비활성화
ax.set(ylabel="Frequency")
ax.set(xlabel="SalePrice")
ax.set(title="SalePrice distribution")

plt.axvline(mu, color='r', linestyle='--') # 새로 점선 추가
plt.text(mu+0.05, 0.111, 'Mean of SalePrice', rotation=0, color='r')
plt.ylim(0, 0.12)
fig.show()
```

로그 변환은 데이터 분포를 정규 분포에 가깝게 만듦

이상치의 영향을 줄여줌 -> 모델 성능 향상
