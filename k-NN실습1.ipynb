```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
# length x
# weight y
# 산점도
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.title('Bream')
plt.show()
```

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.title('smelt & bream')
plt.show()
```

```python
# k-최근접이웃 알고리즘
# 도미와 빙어 데이터 합치기
```

```python
length = bream_length + smelt_length
weigth = bream_weight + smelt_weight
# 이렇게 더하면 가로 방향으로 더해짐더해짐
# 사이킷런 사용하려면 각 특성의 리스트 세로 방향으로 2차원 리스트 만들어야함
```

```python
fish_data = [[l, w] for l, w in zip(length, weigth)]
# zip 함수는 나열된 리스트에서 원소를 하나씩 꺼내줌
# 렝스, 웨잇 하나씩 꺼내서 한 생선을 만들어 준다고 보면 될듯
```

```python
print(fish_data)
# 49개의 생선이 모임
```

```python
# 정답 데이터?
# 첫번째는 도미고 두번째는 뭐고 이런식으로
# 왜?
# 머신러닝 알고리즘이 생선 길이와 무게를 보고 도미와 방어 구분하는 규칙 찾았으면 좋겠음
# 도미와 빙어를 1과 0으로 표현
```

```python
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# 사이킷런패키지에서 k-최근접이웃 알고리즘 구현한 클래스 KNeighborsClassifier 임포트
```

```python
kn = KNeighborsClassifier()
# 임포트한 KNeighborsClassifier 클래스의 객체 만들기
```

```python
kn.fit(fish_data, fish_target)
# fish_data와 fish_target 전달하여 도미 찾는 기준 학습시키기
# 이런 과정을 머신러닝에서 '훈련'이라고 함
# 사이킷런에서는 fit()메서드가 이런 역할을 함

# fit 메서드는 주어진 데이터로 알고리즘 훈련
# 머신러닝 알고리즘을 구현한 프로그램을 모델이라고 부름
```

```python
kn.score(fish_data, fish_target)
# 사이킷런에서 모델 평가하는 메서드는 score() 메서드임
# 메서드는 0에서 1 사이의 값 반환 -> 1은 모든 데이터를 정확히 맞춘 것
```

```python
kn.predict([[30, 600]])
# 사이킷런 라이브러리에서 입력 데이터를 2차원 배열로 처리

# predict() 메서드는 새로운 데이터의 정답 예측
# k-최근접이웃 알고리즘은 어떤 데이터에 대한 답을 구할 때주위 다른 데디터 보고 다수를 차지하는 것을 정답으로 사용
```

```python
# k-최근접이웃알고리즘 -> 데이터를 모두 가지고 있는 것?
# 새로운 데이터에 대해 예측할 때는 가장 가까운 직선거리에 어떤 데이터가 있는지 확인하면 됨
# 하지만 단점은 k-최근접이웃 알고리즘의 이런 특징때문에 데이터가 많은 경우에는 사용하기 어려움
# 데이터가 커서 메모리가 많이 필요하고 직선거리 계산에 시간이 많이 필요함
```

```python
print(kn._fit_X)
# KNeighborsClassifier 클래스는 _fit_X 속성에 fish_data 를 가지고 있음
```

```python
print(kn._y)
# _y 속성에 fish_target 가짐
```

```python
# fit 메서드에 전달한 데이터를 모두 저장하고 있다가 새로운 데이터 나오면 가까운 데이터 참고하여 도미, 빙어 구분
# 가까운 몇 개의 데이터 참고할까? -> 수는 정하기 나름 기본은 5
# 기준은 n_neighbors 매개변수로 바꿈
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
# 참고 데이터를 49개로 한 kn49 모델
# fish_data 의 데이ㅐ터 49개 중 도미가 35개라서 어떤 데이터 넣어도 무조건 도미로 예측
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
print(35/49)
```

k-최근접 이웃 알고리즘


주변에서 가장 가까운 5개의 데이터 보고 다수결 원칙에 따라 데이터 예측


fit(), score(), predict() 메서드 사용

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    kn.n_neighbors = n
    score = kn.score(fish_data, fish_target)

    if score < 1:
        print(n, score)
        break
```
